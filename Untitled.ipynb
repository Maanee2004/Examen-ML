souce de données

```python
data = yf.download("AMZN", period="1d", interval="1m")
data.head()
#print(data.tail())
```

Nettoyer les données

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Vérifier les valeurs manquantes
data.isnull().sum()

# Supprimer les doublons si nécessaire
data = data.drop_duplicates()
data=data.dropna()
data.head()
```

L’évolution du cours de l’action au fil du temps

```python
plt.figure(figsize=(9,6))
plt.plot(data['Close'], label="Cours de clôture AMZN")
plt.title("Évolution du cours Amazon")
plt.xlabel("Date")
plt.ylabel("Prix (USD)")
plt.legend()
plt.show()
```

Visualiser l’évolution du volume de transactions de l’action au fil du temps

```python
plt.figure(figsize=(8,6))
plt.plot(data['Volume'], color="green")
plt.title("Volume de transactions Amazon")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.show()
```

Quelle était la moyenne mobile des différentes actions 

```python
data['MA20'] = data['Close'].rolling(20).mean()
data['MA50'] = data['Close'].rolling(50).mean()

plt.figure(figsize=(12,6))
plt.plot(data['Close'], label="Cours")
plt.plot(data['MA20'], label="MA20", color="red")
plt.plot(data['MA50'], label="MA50", color="green")
plt.title("Moyennes mobiles Amazon")
plt.legend()
plt.show()
```

Quel était le rendement quotidien moyen de l’action

```python
data['Return'] = data['Close'].pct_change()
print("Rendement quotidien moyen AMZN :", data['Return'].mean())
```

Ajouter une nouvelle colonne « Tendance » dont les valeurs 
sont basées sur le « Rendement quotidien »

```python
data['Tendance'] = data['Return'].apply(lambda x: 'Hausse' if x > 0 else 'Baisse')
data.head()
```

Visualiser la fréquence de la tendance à l’aide d’un graphique circulaire 

```python
data['Tendance'].value_counts().plot.pie(autopct='%1.1f%%', figsize=(6,6))
plt.title("Fréquence des tendances Amazon")
plt.show()
```

la corrélation entre les rendements quotidiens des 
différentes actions(APPL,AMZN,MSFT,GOOG)

```python
tickers = ['AMZN','AAPL','MSFT','GOOG']
multi_data = yf.download(tickers, start="2020-01-01", end="2025-01-01")['Close']

# Calcul des rendements quotidiens
returns = multi_data.pct_change()

# Matrice de corrélation
corr = returns.corr()
print(corr)

# Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Corrélation entre les rendements quotidiens")
plt.show()
```
